# IPL Data Analysis Project

This project focuses on analyzing IPL match data to uncover patterns, trends, and insights using data analysis techniques.

Objectives:
- Identify most successful teams
- Analyze impact of toss
- Evaluate player performance

```python
# Import dataset 
import pandas as pd

matches = pd.read_csv("matches (1).csv")
deliveries = pd.read_csv("deliveries (2).csv")

matches.head()
```

```python
# Imformation regarding data

print(matches.shape)
print(matches.columns)
print(matches.info())
print(matches.describe())
```

```python
# Check missing values

matches.isnull().sum()

# Fill or drop missing values
matches['winner'] = matches['winner'].fillna("No Result")
```

```python
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# Most Successful Teams 
import matplotlib.pyplot as plt
wins = matches['winner'].value_counts()
wins.plot(kind='bar', figsize=(12,6))
plt.title("Most Successful IPL Teams", fontsize=15)
plt.xlabel("Teams")
plt.ylabel("Number of Wins")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Toss Impact

toss_win_match_win = matches[matches['toss_winner'] == matches['winner']]
print(len(toss_win_match_win) / len(matches))
```

```python
# Top Players 

top_players = matches['player_of_match'].value_counts().head(10)

top_players.plot(kind='bar', figsize=(10,5))
plt.title("Top Players (Man of the Match)")
plt.show()
```

```python
# Matches per session 

matches['season'].value_counts().sort_index().plot(kind='line')
plt.title("Matches per Season")
plt.show()
```

```python
# Win percentage per Team

total_matches = matches['team1'].value_counts() + matches['team2'].value_counts()
wins = matches['winner'].value_counts()

win_percentage = (wins / total_matches) * 100
win_percentage.sort_values(ascending=False).head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top Teams by Win Percentage")
plt.show()
```

## Final Conclusion

- Mumbai Indians have the highest number of wins (~120+ matches)  
- Toss winner wins approximately 52–55% of matches 
- Certain players consistently outperform others  

This analysis demonstrates how data can be used to extract meaningful insights from sports datasets.
